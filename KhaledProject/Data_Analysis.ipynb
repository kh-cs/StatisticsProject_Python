<h2><b>ðŸŸ¦The Average of Groups Which Passed all Subjects :

```python
d['passing']=((d['math score']>=50) & (d['reading score']>=50) & (d['writing score']>=50))
print(d['passing'].count())
```

<h2><b>ðŸŸ¦The Average of Success and Fail :

```python
size = d['passing'].value_counts()
colors = ['#66b3ff','#ff6666']
labels = "pass", "fail"
explode = [0,0.3]
plt.figure(figsize=(5, 5))
plt.subplot(1,1,1)
plt.pie(size, colors = colors, labels = labels, autopct = '%.2f%%', explode = explode, shadow = True)
plt.title('Groups Success and Fail Average :', fontsize = 20)
plt.legend()#legend for map kay
```

<h2><b>ðŸŸ¦Create Precentage , Grades Symbols ,and Qualification columns  :

```python
d['Percentage'] = (d['math score']+d['reading score']+d['writing score'])/3
def Grade(Percentage):
    if (Percentage >= 95):return 'O'
    if (Percentage >= 81):return 'A'
    if (Percentage >= 71):return 'B'
    if (Percentage >= 61):return 'C'
    if (Percentage >= 51):return 'D'
    if (Percentage >= 41):return 'E'
    else: return 'F'
    
d["grade"] = d.apply(lambda x : Grade(x["Percentage"]), axis=1)

def Qualification(Grade):
    if (Grade == 'O'):return 'Outstanding'
    if (Grade == 'A'):return 'Very Good'
    if (Grade == 'B'):return 'Good'
    if (Grade == 'c'):return 'Average'
    if (Grade == 'D'):return 'Sufficient'
    if (Grade == 'E'):return 'Passable'
    else: return 'Failed'
    
d["Qualification"] = d.apply(lambda x : Qualification(x["grade"]), axis=1)
```
